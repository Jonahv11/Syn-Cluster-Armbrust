## Sorting CMAP Data

```python
# get dataframe of all variables measured in gradients cruises

cruises = ['KOK1606', 'MGL1704', 'KM1906']

# instantiate empty dataframe
catalog_df = pd.DataFrame()
# iterate through cruises
for cruise in cruises:
    # pull catalog of all variables connected to each cruise
    df = api.search_catalog(cruise)
    #check to see if the catalog_df dataframe is empty
    if len(catalog_df) == 0:
        #if it is empty, set catalog_df to the result of the first api search result (first loop)
        catalog_df = df
    else:
        #if it isn't empty, stitch the result of api search (other loops) onto the end of the previously constructed dataframe
        catalog_df = pd.concat([catalog_df, df])
# reset index of concatenated dataframe
catalog_df = catalog_df.reset_index(drop=True)

catalog_df

#this dataframe is a fixed file of all the variables in the three gradients cruises
```

```python
# look how many unique variables we have, count their frequency

# count how many times each variable shows up in catalog_df
var_df = catalog_df['Variable'].value_counts().reset_index()

var_df
```
